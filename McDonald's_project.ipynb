```python
import pandas as pd
df=pd.read_csv(r"/content/menu.csv")
print(df)
```

```python
print(df.head(12)) #top rows
print(df.tail(12))   #bottom rows
```

```python
print(df.info())  #quick way to look datatypes,missing values,data size
```

```python
print(df.shape)   #no.of rows and columns
print(df.shape[0])  #no.of rows
print(df.shape[1])  #no.of columns
```

```python
print(df.columns)   #column names
print(list(df.columns))   #convert to list
```

```python
#checking for missing values
print(df.isnull().head(15))
print(df.isna().sum())   #null values & adding
```

```python
##isolating columns

print(df['Category'].head(15))   #one column
print(df[['Item','Calories']].head(15))   #multiple column

##isolating rows
print(df[df.index==15])
```

```python
#using.loc to fetch rows
print(df.loc[20])
print(df.loc[20:26])
print(df.iloc[10])
print(df.iloc[18:22])
```

```python
#dealing with missing values
##dropping missing values
data1=df.dropna()
print(data1.isnull().sum())
df.dropna(inplace=True,axis=0)
print(df.isnull().sum())
```

```python
#replacing mean
mean_value=df['Total Fat'].mean()
print(mean_value)
#replacing median
median_value=df['Carbohydrates'].median()
print(median_value)
```

```python
#renaming column
df.rename(columns={'Calories':'CALORIES'},inplace=True)
df.head(20)
```

```python
#value.counts()
df['Sodium'].value_counts()
```
