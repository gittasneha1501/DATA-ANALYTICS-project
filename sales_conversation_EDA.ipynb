```python
#checking for missing values
print(df.isnull().head(15))
print(df.isna().sum())   #null values & adding
```

```python
#slicing & extracting data

##isolating columns

print(df['interest'].head(10))    #one column
print(df[['Impressions','age']].head(15))   #multiple column

##isolating rows
print(df.index==20)
```
