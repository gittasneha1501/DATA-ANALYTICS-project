```python
import pandas as pd
df=pd.read_csv(r"/content/Advertising.csv")
print(df)
```

```python
print(df.head(15)) #top rows
print(df.tail(15))   #bottom rows
```

```python
print(df.info())  #quick way to look datatypes,missing values,data size
```

```python
print(df.shape)   #no.of rows and columns
print(df.shape[0])  #no.of rows
print(df.shape[1])  #no.of columns
```

```python
print(df.columns)   #column names
print(list(df.columns))   #convert to list
```

```python
#checking for missing values
print(df.isnull().head(10))
print(df.isna().sum())   #null values & adding
```

```python
#slicing & extracting data

##isolating columns

print(df['Radio'].head(20))    #one column
print(df[['Newspaper','Sales']].head(20))   #multiple column

##isolating rows
print(df.index==15)
```

```python
#using.loc to fetch rows
print(df.loc[12])
print(df.loc[9:16])
print(df.iloc[5])
print(df.iloc[8:12])
```

```python
#dealing with missing values
##dropping missing values
data1=df.dropna()
print(data1.isnull().sum())
df.dropna(inplace=True,axis=0)
print(df.isnull().sum())
```

```python
#replacing mean
mean_value=df['TV'].mean()
print(mean_value)
#replacing median
median_value=df['Radio'].median()
print(median_value)
```
