```python
import pandas as pd
df=pd.read_csv(r"/content/zomato.csv")
print(df)
```

```python
print(df.head(25)) #top rows
print(df.tail(25))   #bottom rows
```

```python
print(df.info())  #quick way to look datatypes,missing values,data size
```

```python
print(df.shape)   #no.of rows and columns
print(df.shape[0])  #no.of rows
print(df.shape[1])  #no.of columns
```

```python
print(df.columns)   #column names
print(list(df.columns))   #convert to list
```

```python
#checking for missing values
print(df.isnull().head(20))
print(df.isna().sum())   #null values & adding
```

```python
#slicing & extracting data

##isolating columns

print(df['restaurant type'].head(10))    #one column
print(df[['cuisines type','rate (out of 5)']].head(10))   #multiple column

##isolating rows
print(df.index==25)
```

```python
#using.loc to fetch rows
print(df.loc[15])
print(df.loc[12:18])
print(df.iloc[9])
print(df.iloc[6:14])
```

```python
#dealing with missing values
##dropping missing values
data1=df.dropna()
print(data1.isnull().sum())
df.dropna(inplace=True,axis=0)
print(df.isnull().sum())
```

```python
#replacing mean
mean_value=df['avg cost (two people)'].mean()
print(mean_value)
#replacing median
median_value=df['num of ratings'].median()
print(median_value)
```

```python
#renaming column
df.rename(columns={'avg cost (two people)':'avg cost'},inplace=True)
df.head(15)
```
